disini kita menampilkan 10 record pertama dataset dari atribute bertipe angka

```python
data = {}

data['Jumlah Record'] = ibnu.shape[0]
data['Jumlah Attribute'] = ibnu.shape[1]
data['Nama Attribute'] = ibnu.columns.tolist()
data['Jumlah Attribute Dengan Tipe Angka'] = ibnu.select_dtypes(include=['int64', 'float64']).shape[1]

dictKeys = list(data.keys())

table = prtl.PrettyTable()
table.add_column(dictKeys[0], [data['Jumlah Record']])
table.add_column(dictKeys[1], [data['Jumlah Attribute']])
table.add_column(dictKeys[2], ['\n'.join(data['Nama Attribute'])])
table.add_column(dictKeys[3], [data['Jumlah Attribute Dengan Tipe Angka']])
print(table)
```

menampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka.
